```python
import pandas as pd
import os

# Path to the folder containing images
images_folder = "Images"  # Path where your images are stored
csv_file = "GTruth.csv"  # Path to your GTruth.csv file

# Load the ground truth CSV file
gtruth = pd.read_csv(csv_file)

# Get the list of image files in the folder (without file extension)
image_files = [os.path.splitext(f)[0] for f in os.listdir(images_folder) if f.endswith(('.jpeg', '.jpg', '.png'))]

# Ensure both the image filenames and the Ids are strings to avoid format mismatches
gtruth["Id"] = gtruth["Id"].astype(str)  # Convert IDs to string for comparison

# Print out the first few image filenames to inspect
print("Image files found:", image_files[:10])  # Show first 10 files for checking

# Print out the first few 'Id' values from GTruth to inspect
print("GTruth IDs:", gtruth["Id"].head())

# Filter the rows in GTruth.csv where the image exists in the folder
gtruth_filtered = gtruth[gtruth["Id"].isin(image_files)]

# Print out the shape of the filtered DataFrame to check how many rows remain
print("Filtered DataFrame shape:", gtruth_filtered.shape)

# Save the filtered DataFrame to a new CSV file
output_file = "/content/Filtered_GTruth.csv"  # Path to save the cleaned CSV
gtruth_filtered.to_csv(output_file, index=False)

# Print confirmation
print(f"Filtered CSV saved to: {output_file}")
```
